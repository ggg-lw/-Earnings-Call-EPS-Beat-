```python
import pandas as pd
from sklearn.metrics import log_loss, brier_score_loss

# 1. 读取数据
file_path = '/Users/xinyuewang/Desktop/1.27/data/beat_probability_results.csv'
df = pd.read_csv(file_path)

# 2. 定义计算函数
def calculate_metrics(group):
    # 计算 Log Loss (对数损失)
    ll = log_loss(group['actual_beat'], group['predicted_beat_prob'], labels=[0, 1])
    # 计算 Brier Score (布里尔分数)
    bs = brier_score_loss(group['actual_beat'], group['predicted_beat_prob'])
    return pd.Series({'log_loss': ll, 'brier_score': bs})

# 3. 分组统计
summary = df.groupby('is_out_of_sample').apply(calculate_metrics).reset_index()

# 4. 将结果合并回原始 DataFrame
df = df.merge(summary, on='is_out_of_sample', how='left')
df.rename(columns={
    'log_loss': 'sample_type_log_loss', 
    'brier_score': 'sample_type_brier_score'
}, inplace=True)

# 5. 保存回 CSV 文件
df.to_csv(file_path, index=False)
print("统计完成，结果已存入原文件。")
```
