```python
import yfinance as yf
import pandas as pd


def format_quarter(dt):
    p = pd.to_datetime(dt).to_period("Q")
    return f"{p.year}-Q{p.quarter}"


def fetch_earnings_from_yahoo(ticker: str):
    print(f"抓取 {ticker} from Yahoo Finance")

    tk = yf.Ticker(ticker)
    earnings = tk.earnings_dates

    if earnings is None or earnings.empty:
        print(f"⚠️ {ticker}: 没有 earnings 数据")
        return None

    df = earnings.reset_index()

    # 统一列名
    df = df.rename(columns={
        "Earnings Date": "announcement_date",
        "EPS Estimate": "estimated_eps",
        "Reported EPS": "actual_eps"
    })

    df["fiscal_quarter"] = df["announcement_date"].apply(format_quarter)
    df["ticker"] = ticker.upper()

    df = df[[
        "ticker",
        "announcement_date",
        "fiscal_quarter",
        "estimated_eps",
        "actual_eps"
    ]].copy()

    print(f"✅ {ticker}: {len(df)} 条 earnings")

    return df
```

```python
df["beat"] = (df["actual_eps"] > df["estimated_eps"]).astype(int)
```
